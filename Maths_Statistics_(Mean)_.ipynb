#MEDIAN


```python
data = [5, 2, 9, 1, 7]
data.sort()
n = len(data)
median = data[n//2] if n % 2 != 0 else (data[n//2 - 1] + data[n//2]) / 2
print("Median:", median)
```

#MODE


```python
n = int(input("Enter number of elements: "))
data = []
for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    data.append(num)

mode = data[0]
max_count = 0
for num in data:
    count = 0
    for x in data:
        if x == num:
            count += 1
    if count > max_count:
        max_count = count
        mode = num

print("Mode:", mode)
```
