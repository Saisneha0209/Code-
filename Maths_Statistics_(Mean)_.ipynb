```python
#Mean
sum=0
n=int(input("Enter how many numbers to be entered:"))
for i in range(n):
  number=int(input(f"Enter nummber {i+1} :"))
  sum=sum+number
print(sum)
mean=sum/n
print("So,The mean of your given numbers is",mean)
```
