```python
import numpy as np
matrix_a=np.array([[5,6],
                   [7,8]])
matrix_b=np.array([[3,8],
                   [7,6]])
print(matrix_a + matrix_b)
```

```python
matrix_a=([[5,6],
          [7,8]])
matrix_b=([[3,8],
          [7,6]])

matrix_c = [[0,0],[0,0]]

for i in range(len(matrix_a)):
  for j in range(len(matrix_a[0])):
    matrix_c[i][j] = matrix_a[i][j] + matrix_b[i][j]
print("THE RESULT OF ADDITION IS:")
for row in matrix_c:
  print(row)
```
