```python
#to calculate simple interest here:-
class SimpleInterest:
  def __init__(self):
    self.p=int(input("enter amount"))
    self.r=int(input("enter rate"))
    self.t=int(input("enter time"))



class Calculation(SimpleInterest):
  def cal(self):
    si=(self.p*self.r*self.t/100)
    print("SI is",si)

obj=Calculation()
obj.cal()
```

#MULTI LEVEL INHERITANCE


```python
class A:#parent class
  def info(self):
    print("Class A")
class B(A):      #child class
  def info1(self):
    print("Class B")
class C(B):     #child class of CLASS B
  def info2(self):
    print("Class C")


obj=C()
obj.info()
obj.info1()
obj.info2()
```

#POLYMORPHISM 1.function overiding and function overloading


```python
#FUNCTION OVERRIDING:-
class A:#parent class
  def info(self):
    print("Class A")
class B(A):      #child class
  def info(self):
    super().info() #this function will print class A and class B both
    print("Class B")

obj=B()
obj.info()
```

```python
class A:#parent class
  def info(self):
    print("Class A")
class B(A):      #child class
  def info(self):
    super().info() #this function will print class A and class B both #binding of data
    print("Class B")
class C(B):
  def info(self):
    super().info()
    print("Class C")

obj=C()
obj.info()
```

#Encapsulations

```python
class A: #variables of data is hidden#data is hidden
  def __init__(self,age):
    self._age=age

class B(A):
  def age1(self):
    print(self._age)

obj=B(20)
obj.age1()
```
