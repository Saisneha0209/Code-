```python
#INHERITANCE(PRODUCT KE FUNCTIONS KE RESULT IN ORDER WISE)
#parent class is product and child class is order
#single level inheritance
class Employee:
  def info(self):
    print("Employee name is Saisneha")

class Department(Employee):#for child class inherited parent class name
  def dept(self):
    print("Department is Electrical")

obj=Department()
obj.info()
obj.dept()
```
