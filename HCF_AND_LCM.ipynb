```python
num1 = 15
num2 = 20

hcf = 1
for i in range(1,num1 + 1):
  if num1%i == 0 and num2%i == 0:
    hcf = i

print(hcf)
```

```python
num1 = 15
num2 = 20

hcf = 1
for i in range(1,num1 + 1):
  if num1%i == 0 and num2%i == 0:
    hcf = i
lcm = (num1*num2)//hcf
print(lcm)
```
