```python
import pandas as pd
from sklearn.linear_model import LinearRegression  #80% is there to train it and the rest of the 20% to be given to the testing data frame...
from sklearn.model_selection import train_test_split #classes
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('house_data_large (1).csv')
#print(df)
x=df[["Sqft","Bedrooms","Bathrooms","YearBuilt","LocationCode"]]
y=df["Price"]
#print(x)
#print(y)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
#print(pred)#TO FIND THE PREDICTION....
user_input=np.array([[3692,1,1,2015,5]])
u_pred=model.predict(user_input)
print(int(u_pred))
```

```python
#to show it diagramatically
import matplotlib.pyplot as plt
df = pd.read_csv('house_data_large (1).csv')
print(df)
x=df[["Sqft","Bedrooms","Bathrooms","YearBuilt","LocationCode"]]
y=df["Price"]
print(x)
print(y)
plt.plot(x,y)#Here it will be line graph..
plt.title("Representation of houses for information")
plt.xlabel('Sqft')
plt.ylabel('Price')
plt.show()


```
