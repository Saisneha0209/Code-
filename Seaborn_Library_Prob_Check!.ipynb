#Using NUMPY(Random Module) to check the probability and using SEABORN to visualize them!!

```python
#Seaborn Library!!
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import random

#used when only single column
a = random.normal(loc=2, scale = 1,size=3)
print(a)
sns.kdeplot(random.normal(loc=2, scale = 1,size=3))
plt.show()

#binomial distribution - based on 2-it show what happen (Discret)
#size -shape of array,n-no. of trails,p-probability
b = random.binomial(n=5,p=0.5,size=4)
print(b)
sns.histplot(random.binomial(n=5,p=0.5,size=4), kde=True)
plt.show()

# poison distribution - rate of occurence -It predict time (Discret)
#size - shape of array,lam- Rate or non occurence

# uniform distribution -
# only have size
# important -Logistict distribution-discribe the growth ,used in logistic(ML),neural
# same parameter(size ,scale,loc),used when multiple column are check.
# mutlinomial distribution -check the probabilty  more than 2 (n- no. of time,pvals-list of probability,size)

b = random.multinomial(n=3,pvals=[1/3,1/3,1/3],size=10)
print(b)

# Exponational distribution -Inverse of lam- Faliure and pass (Scale(inverse of lam) ,size).
# chisquare distribution- assumption verify . (size,df-(dgree of freedom))
# Rayleigh distribution - how much  signal come in your route (size-shape of array,scale-SD)
# Perito distribution - depend on outcome ,independent about factor(a-shape of parameter,size)
#Zipf distribution - frequency(comman ) - (a- distribution of parameter,size)
```
