```python
# Step 1: Import necessary libraries
import pandas as pd
from google.colab import files

# Step 2: Upload any file from your computer
uploaded = files.upload()

# Step 3: Get the uploaded file name
for filename in uploaded.keys():
    print(f"\n File uploaded successfully: {filename}")
    file_name = filename  # store the file name

# Step 4: Detect file type and read accordingly
if file_name.endswith('.csv'):
    df = pd.read_csv(file_name)
elif file_name.endswith(('.xls', '.xlsx')):
    df = pd.read_excel(file_name)
else:
    raise ValueError("Unsupported file type! Please upload a CSV or Excel file.")

# Step 5: Display the first few rows of the DataFrame
print("\nDataFrame Preview:")
print(df.head())

# Step 6: (Optional) View DataFrame info
print("\nℹ DataFrame Info:")
print(df.info())
```
