```python
#Permutation
def factorial(n):
  result = 1
  for i in range(1,n+1):
    result *= i
  return result

def npr(n,r):
  if r>n:
      return 0
  return factorial(n) //factorial(n-r)

def ncr(n,r):
  if r>n:
      return 0
  return factorial(n) // (factorial(r) * factorial(n-r))


n=5
r=3
print(f"Total (n):{n}")
print(f"To choose (r):{r}")

print(f"PERMUTATION (NPR):{npr(n,r)}")
print(f"COMBINATION (NCR):{ncr(n,r)}")
```
