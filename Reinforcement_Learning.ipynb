#REINFORCEMENT LEARNING!!

```python
reward = 0
for num in range(1,101):
  if num%2==0:
    reward+=1
print("Reward",reward)
```

```python
#Random value:-
import random
rl = {"A":0,"B":0}
for i in range(100):
  turn = random.choice(["A","B"])
print(turn)
reward=1 if turn=="A" else 0
rl[turn]+=reward
print(rl)
```

```python
#next level
rl = {"A":0,"B":0}
for i in range(100):
   turn = "A" if rl["A"]>=rl["B"] else "B"
print(turn)
reward=1 if turn=="A" else 0
rl[turn]+=reward
print(rl)
```

```python
#next step
import random
rl = {"A":0,"B":0}
choclate = 0
count = 0
while choclate<100:
   turn = "A" if rl["A"]>=rl["B"] else "B"
   print(turn)
   reward=1 if turn=="A" else 0
   rl[turn]+=reward
   choclate+=reward
   count = count+1
   print("Goal reached")
   print(count)
   print(rl)
```

```python
#next step
import random
rl = {"A":-10,"B":-5}
choclate = 0
count = 0
while choclate<100:
  turn = "A" if rl["A"]>=rl["B"] else "B"
  print(turn)
  reward=1 if turn=="A" else -1
  rl[turn]+=reward
  choclate+=reward
  count = count+1
  print("Goal reached")
  print(count)
  print("rl",rl)
```
