```python
import random
rl={"A":0,"B":0}
for i in range(100):
  turn=random.choice(["A","B"])
  reward=1 if turn=="A" else 0
  rl[turn]+=reward
print(rl)
```

```python
import random
rl={"A":0,"B":0}
chocolate=1
count=0
while chocolate<100:
    turn="A" if rl["A"]>=rl["B"] else "B"
    reward=1 if turn=="A" else 0
    rl[turn]+=reward
    chocolate +=reward
    count=count+1
    if count % 10 == 0:
        print(f"Chocolate: {chocolate}, Rewards: {rl}")
print("Goal Reached")
print(count)
print(rl)
```
