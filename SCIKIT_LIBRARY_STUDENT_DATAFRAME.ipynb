#PROGRAMMING THE MODEL THAT WILL CHECK OUT-student will pass or fail(not exactly but likely)!!

```python
import pandas as pd  #import panda library
from sklearn.linear_model import LogisticRegression #here logistic regression is used to make a decision...
from sklearn.model_selection import train_test_split
import numpy as np  #without numpy also you will get your desired output...
import warnings
warnings.filterwarnings('ignore')  #to avoid the warning you get after making this run...
df = pd.read_csv('students_1000.csv')
x=df[["study_hours","sleep_hours"]]
y=df["pass"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=30)
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
user_input_pass=np.array([[7.53,7.95]])
pred_pass=model.predict(user_input_pass)
#print("With the given info the student will:",(int(pred_pass))),if you want to check manully you can do this..
#wanna get the info from the user itself check this out..
study_hours = int(input("ENTER STUDY HOURS : "))
sleeping_hours = int(input("ENTER SLEEPING HOURS : "))
stud_info = [[study_hours, sleeping_hours]]
pred = model.predict(stud_info)
if sleeping_hours <=10:
  if pred[0] == 1:
     print("Student is likely to pass..Well done!")
  else:
     print("Student is likely to failed...Need improvement!Work more hard next time!!")
else:
   print("Reduce your sleeping hours and try to increase your studying hours!!")
print("HENCE,You got your information keep doing more in your field,you will soon get more result soon!!")
```
