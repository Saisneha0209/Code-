#Single Inheritance:-

```python
#This is the Single Inheritance:-
class Items:
    def available_items(self):
        print("Items available:5")
        print("-------------------------")
    def first_item(self):
        print("First Item", "Sandwich")
    def second_item(self):
        print("Second Item", "Pizza")
    def third_item(self):
        print("Third Item", "Shake")
        print("-------------------------")
    def first_item_price(self):
        print("Price for the first item:365")
    def second_item_price(self):
        print("Price for the second item:150")
    def third_item_price(self):
        print("Price for the third item:290")
        print("-------------------------")
class Users(Items):
    def user_order(self):
        print("Order:", "Pizza")
    def user_bill(self):
        print("User bill:", 150)

object = Users()
object.available_items()
object.first_item()
object.second_item()
object.third_item()
object.first_item_price()
object.second_item_price()
object.third_item_price()
object.user_order()
object.user_bill()
```

#Multiple Inheritance:-

```python
# This is the Multiple Inheritance:-
class Items:
    def item1(self):
        print("Item1", "Sandwich")
    def item2(self):
        print("Item2", "Burger")
    def item3(self):
        print("Item3", "Coffee")
        print("---------------------------")
    def item1_price(self):
        return 300
    def item2_price(self):
        return 70
    def item3_price(self):
        return 30

class Order:
    def order1(self):
        print("Order-1", "Pizza")
    def order2(self):
        print("Order-2", "Sandwich")
    def order3(self):
        print("Order-3", "Noodles")
        print("----------------------------")
class Bill(Items, Order):
    def bill(self):
      user_bill=0
      Item1_price=250
      Item2_price=450
      Item3_price=355
      user_bill = (user_bill+Item1_price+Item2_price+Item3_price)
      print("Final Bill Is:", user_bill)
      print("---------------------------")
object = Bill()
object.item1()
object.item2()
object.item3()
print("Item-1 price is:", object.item1_price())
print("Item-2 price is:", object.item2_price())
print("Item-3 price is:", object.item3_price())
object.order1()
object.order2()
object.order3()
object.bill()
```

#Heirarchical Inheritance:-

```python
#This is the Heirarchical Inheritance:-
class Items:
    def item1(self):
        print("Item-1:", "Sandwich")
    def item2(self):
        print("Item-2:", "Burger")
    def item3(self):
        print("Item-3:", "Chocolate Shake")
        print("|------------------------ |")
    def item1_price(self):
        print("Item-1 price:", 270)
        return 270
    def item2_price(self):
        print("Item-2 price:", 250)
        return 250
    def item3_price(self):
        print("Item-3 price:", 120)
        print("|------------------------|")
        return 120

class User1(Items):
    def order1(self):
        print("Order-1:", "Chocolate Shake")

class User2(Items):
    def order2(self):
        print("Order-2:", "Pizza")
        print("|---------------------------|")

class Bill(Items):
    def bill(self):
        user_bill = 0
        item1_price =270
        item2_price =250
        item3_price =120
        user_bill = user_bill + item1_price + item2_price + item3_price
        print("Final Bill Is:", user_bill)
        print("|---------------------------|")
#First call the objects!!
object1 = User1()
object2 = User2()
object3 = Bill()

object1.item1()
object1.item2()
object1.item3()

object1.order1()
object2.order2()
#Final Bill:-
object3.bill()
```

#Polymorphism:-

```python
#Function Overriding!
class A:
    def info(self):
      super().info()
      print("Hello!")
class B(A):
    def info(self):
      print("Hii there!")

obj = B()
obj.info()
```

#Multi Level Inheritance!

```python
class a:
    def info(self):
        print("Hello!")
class b(a):
    def info(self):
      print("Hii!")
class c(b):
    def info(self):
      super().info()
      print("Hey there!")
obj = c()
obj.info()
```

```python
#Here heirarchical and multiple together it makes HYBRID!!
class A:
    def info(self):
        print("Hii A here!")
class B(A):
    def info(self):
      super().info()
      print("Hii B here!")
class C(A):
    def info(self):
      super().info()
      print("Hii C here!")
class D(B, C):
    def info(self):
      super().info()
      print("Hii D here!")
obj = D()
obj.info()
```

#ENCAPSULATIONS(To wrap the data/moree use in security!!)

```python
class A:
    def __init__(self, a, b):
        self.__a = a
        self.b = b
        self.c = a + b

class B(A):
    def info(self):
      print(self.__a)
      print(self.c)

obj = A(5, 3)
obj.info()
```

#Abstraction:-help hiding the code with security-type!!

```python
from abc import ABC, abstractmethod
class Payment(ABC):
    @abstractmethod
    def payment(self, method, amount):
        pass

class UpiPayment(Payment):
    def payment(self, method, amount):
        print(f"{amount} done by {method} done")

class CreditCardPayment(Payment):
    def payment(self, method, amount):
        print(f"{amount} done by {method} done")

class DebitCardPayment(Payment):
    def payment(self, method, amount):
        print(f"{amount} done by {method} done")

def payment_process(method):
    if method.lower() == "upi":
        return UpiPayment()
    elif method.lower() == "credit":
        return CreditCardPayment()
    else:
        return DebitCardPayment()
pay_object = payment_process("UPI")
pay_object.payment("Gpay", 10000)
```
